```python
# Importing necessary libraries
import pandas as pd
import numpy as np
```

```python
#Loading the dataset
data = pd.read_csv(r'Downloads/messy_HR_data.csv')
data
```

```python
# Removing Extra Spaces from all columns
for col in data.columns:
    if data[col].dtype == 'object':
        data[col] = data[col].str.strip()   
```

```python
# Capitalizing all first letters in the Name Column
data['Name'] = data['Name'].str.title()
```

```python
# Handling the Age Column
# Replacing thirty with 30
data['Age'] = data['Age'].replace('thirty', 30)

#Converting All Age to numeric
data['Age'] = pd.to_numeric(data['Age'], errors = 'coerce')

#Filling missing values in Age column
data['Age'] = data['Age'].fillna(data['Age'].median())

# Converting the Age to int
data['Age'] = data['Age'].astype(int)
data
```

```python
# Handling the Salary Column
# Replacing SIXTY THOUSAND with 60000
data['Salary'] = data['Salary'].replace('SIXTY THOUSAND', 60000)

# Converting Salary to numeric
data['Salary'] = pd.to_numeric(data['Salary'], errors = 'coerce')

# Filling missing values in the Salary column
data['Salary'] = data['Salary'].fillna(data['Salary'].median())

# Converting the Salary to int
data['Salary'] = data['Salary'].astype(int)
data
```

```python
# Cleaning the Gender Column
data['Gender'] = data['Gender'].str.title()
```

```python
#Defining a new function into parse_date
def parse_date(x):
    try:
        return parser.parse(str(x).strip())
    except:
        return pd.NaT

data['Joining Date'] = data['Joining Date'].apply(parse_date)
data['Joining Date'] = pd.to_datetime(data['Joining Date'])
data
```

```python
# Filling Missing Values in the Email Column with'Missing"
data['Email'] = data['Email'].fillna('Missing')
data
```

```python
# Filling Missing Values in the Phone Number Column with'Missing"
data['Phone Number'] = data['Phone Number'].replace(r'^\s*$', np.nan, regex=True) #Some cells hae white spaces
data['Phone Number'] = data['Phone Number'].fillna('Missing')
data
```

```python
#Filtering Out the Staff with Missing Email or Phone Number

email_missing_mask = data['Email'] == 'Missing' # This creates a mask for the emails with missing fields

pn_missing_mask = data['Phone Number'] == 'Missing' # This creates a mask for the phone number with missing fields

missing_details = email_missing_mask | pn_missing_mask # This combines the mask with the OR (|) Operator

staff_with_missing_details = data[missing_details] # This filters the dataframe to select only the data with missing email or phone number

staff_with_missing_details 
```
